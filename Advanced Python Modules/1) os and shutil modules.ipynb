## <code>os.walk()</code>
Access every sub-directory and files present inside parent directory in tree like manner!

```python
dir_path = os.getcwd() + '\\Top_Level_One'
dir_path
```

```python
# Printing every file and sub-folder name present in Top_Level_One
for folder, sub_folders, files in os.walk(dir_path):
    print(f'Currently looking at --> {folder}')
    
    print('Sub-folders are -->')
    i = 1
    for sub_folder in sub_folders:
        print(f'\t{i}) {sub_folder}')
        i += 1

    print('Files are -->')
    i = 1
    for file in files:
        print(f'\t{i}) {file}')
        i += 1
        
    print('\n')
```

```python
# Accessing and reading .txt files! We will get error because os.walk() just mentions the name of directory
# path which it is currently looking at, it doesn't actually change directory from the dir we are currently
# in to the dir we are looking at.

for folder, sub_folders, files in os.walk('C:\\Users\\YasH\\Advanced Python Modules\\Top_Level_One'):
    print(f'Currently in --> {folder}')

    i = 1
    for file in files:
        f = open(file,'r')
        f.seek(0)
        print(f'\t{i}) {file} has text --> {f.read()}')
        f.close()
        i += 1
        
    print('\n')
```

```python
# Accessing and reading .txt files! Here we actually have to provide the full path of the files since the files are not
# in our current working dir

for folder, sub_folders, files in os.walk('C:\\Users\\YasH\\Advanced Python Modules\\Top_Level_One'):
    print(f'Currently in --> {folder}')

    i = 1
    for file in files:
        file_path = folder + f'\\{file}'
        f = open(file_path,'r')
        f.seek(0)
        print(f'\t{i}) {file} --> {f.read()}')
        f.close()
        i += 1
        
    print('\n')
```

```python
# Printing all the sub-folders and their contents in Desktop

for folder, sub_folders, files in os.walk('C:\\Users\\Yash Happy Birthday\\Desktop'):
    print(f'Currently in --> {folder}')
    
    print('SUB-FOLDERS -->')
    i = 1
    for sub_folder in sub_folders:
        print(f'\t{i}) {sub_folder}')
        
    print('FILES -->')
    i = 1
    for file in files:
        print(f'\t{i}) {file}')
        
    print('\n')
```
